```python
query = {'project_id': project_id}

cursor = activity.find(query).sort("index", pymongo.ASCENDING)

activity_df = pd.DataFrame(list(cursor))
```

```python

query = {}

cursor = annotations.find(query)

annotation_df = pd.DataFrame(list(cursor))
```

```python
taxonomy_map = {
    1: "Idea Generation", 2: "Idea Organization", 3: "Discourse Planning",
    4: "Drafting", 5: "Lexical Chaining", 6: "Object Insertion",
    7: "Semantic", 8: "Syntactic", 9: "Lexical",
    10: "Structural", 11: "Visual", 12: "Quantitative",
    13: "Feedback", 0: "No Label", 14: "Artifact", 15:"Command Insertion",
    16: "Citation"
}
```

```python
annotation_array = annotation_df[project_id][0]["filledArray"]
```

```python
for i, e in enumerate(annotation_array):
  print(i, e)
```

```python

def get_annotation_spans(arr):
  low = 0
  new_arr = []

  for i in range(len(arr)):
    if (arr[i] == [0] or (arr[i] == arr[low] and activity_df.iloc[i]["file"] == activity_df.iloc[low]["file"])):
      continue
    else:
      try:
        key = int(arr[low][0]) 
      except:
        continue

      if (key in taxonomy_map):
        new_arr.append((taxonomy_map[key], low, i))
        low = i
  
  try:
    key = int(arr[low][0]) 
  except:
    key = -1

  if (low < len(arr) and key in taxonomy_map):
    new_arr.append((taxonomy_map[key], low, len(arr)))
  return new_arr

annotation_spans = get_annotation_spans(annotation_array)
```

```python
annotation_spans = get_annotation_spans(annotation_array)
```

```python
print(annotation_spans)
```

```python
fine_tuning_db = client["dataset_db"]["fine_tuning"]
fine_tuning_db.delete_many({})

for span in annotation_spans:
  instruction = "INSTRUCTION FOR WRITING INTENTION"
  writing_intention = span[0]
  before_diff_arr= activity_df.iloc[span[1]]["revision"]
  after_diff_arr = activity_df.iloc[span[2]]["revision"]

  before_text = ""
  after_text = ""

  for arr in before_diff_arr:
    key, string = arr[:2]
    if (key == 1):
      continue
      
    before_text += string

  for arr in after_diff_arr:
    key, string = arr[:2]
    if (key == 1):
      continue
      
    after_text += string
  
  fine_tuning_db.insert_one({
    "instruction": instruction,
    "writing_intention": writing_intention,
    "before_text": before_text,
    "after_text": after_text,
    "start_idx": span[1],
    "end_idx": span[2]
  })
  #print("before_text\n", before_text)
  #print("after_text\n", after_text)
```
